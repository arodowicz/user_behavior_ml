
import libraries and data

```python
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('/datasets/users_behavior.csv')
```

```python
print(df.info())
print(df.head())
```

    

Split data into training set, validation set and final test set with a 60/20/20 ratio 



```python
features = df.drop(['is_ultra'], axis=1)
target = df['is_ultra']
```

```python
features_train, features_verify, target_train, target_verify = train_test_split(features, target, test_size=0.4, random_state=12345)
```

```python
features_valid, features_test, target_valid, target_test = train_test_split(features_verify, target_verify, test_size=0.5, random_state=12345)
```


Good splitting of the data into `train`, `validation` and `test` datasets.


```python


best_accuracy = 0
best_depth = None

for depth in range(1,11):
    model = DecisionTreeClassifier(max_depth=depth, random_state=12345)
    model.fit(features_train, target_train)
    predictions = model.predict(features_verify)
    accuracy = accuracy_score(target_verify, predictions)

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_depth = depth

print('Best Accuracy :', best_accuracy)
print('Best Depth :', best_depth)


    
```



   Testing the DecisionTreeClassifier model for best accuracy and depth to compare with others. Best Accuracy was 0.79, which is higher than the minimum requirement. 



```python

best_depth = 0 
best_score = 0
best_est = 0
for est in range(10,51,10):
    for depth in range(1,11):
        model = RandomForestClassifier(random_state=12345, n_estimators=est, max_depth=depth)
        model.fit(features_train, target_train)
        score_valid = model.score(features_verify, target_verify)
        predictions = model.predict(features_verify)
      
        if score_valid > best_score:
            best_score = score_valid
            best_est = est
            best_depth = depth

print('best depth:', best_depth, 'best score:', best_score, 'best est:', best_est)   
```


Testing the RandomForestClassifier model to find best accuracy result. with a 0.803 it is higher than both the minimum requrement as well as the DecisionTreeClassifier Model.


```python

model = LogisticRegression(random_state=12345)
model.fit(features_train, target_train)
train_score = model.score(features_train, target_train)
valid_score = model.score(features_verify, target_verify)
print('accuracy of training model:', train_score * 100)
print('accuracy of validation model:', valid_score * 100)
```



   LogisticRegression Model did not meet minimum requrements.

```python


final_model = RandomForestClassifier(random_state=12345, n_estimators=30, max_depth=7)
final_model.fit(features_train, target_train)

final_score = final_model.score(features_test, target_test)

print('final training score:', final_score)

```

    

 Final Model with the higest accuracy was the RandomForestClassifier. Tested the final data against the training data with a final score of 0.804
